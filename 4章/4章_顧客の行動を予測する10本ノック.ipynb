### ノック32：クラスタリングで顧客をグループ化しよう

```python
customer_clustering = customer[['mean', 'median', 'max', 'min', 'membership_period']]
customer_clustering.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 標準化の準備
sc = StandardScaler()
# フィッティングとデータ変換
customer_clustering_sc = sc.fit_transform(customer_clustering)
# k-meansモデル構築
kmeans = KMeans(n_clusters = 4, random_state = 0)
# モデル学習（クラスタリング）
clusters = kmeans.fit(customer_clustering_sc)
# 各データのクラスタ番号をリスト化
customer_clustering['cluster'] = clusters.labels_
print(customer_clustering['cluster'].value_counts())
customer_clustering.head()
```

### ノック33：クラスタリング結果を分析しよう

```python
# 分析結果を把握しやすくするためにカラム名を変更
customer_clustering.columns=['月内平均値', '月内中央値', '月内最大値', '月内最小値', '会員機関', 'cluster']
# クラスタごとにデータ件数を集計
customer_clustering.groupby('cluster').count()
```

```python
# クラスタごとの平均
customer_clustering.groupby('cluster').mean()
```

### ノック34：クラスタリング結果を可視化してみよう

可視化にあたって、クラスタ変数の5つのカラムを2つに次元削減する必要あり  
今回は主成分分析(PCA)を用いる

```python
from sklearn.decomposition import PCA

# xにクラスタリングした変数を代入
x = customer_clustering_sc
# モデル定義
pca = PCA(n_components = 2) # 二次元
# 学習
pca.fit(x)
# 二次元に変換
x_pca = pca.transform(x)
# 二次元に圧縮したデータをデータフレーム型へ変換
pca_df = pd.DataFrame(x_pca)
# 32でやったクラスタリング結果の格納
pca_df['cluster'] = customer_clustering['cluster']
```

```python
# 可視化していく
import matplotlib.pyplot as plt
%matplotlib inline
# クラスタごとに散布図を描画
for i in customer_clustering['cluster'].unique():
    tmp = pca_df.loc[pca_df['cluster'] == i]
    plt.scatter(tmp[0], tmp[1])
```

### ノック36：翌月の利用回数予測を行うためのデータ準備をしよう  
・教師となる過去データを整える  
・特徴となるデータを生成  
・予測モデルを作成  
・データセットを学習用データと評価用データに分割  
・評価用データでモデルの精度を検証  
・モデルを用いて未知データの予測

```python
# usedateを日時を表す文字列datetime型に変換し、strftimeメソッドで年月の形式にする
uselog['usedate'] = pd.to_datetime(uselog['usedate'])
uselog['年月'] = uselog['usedate'].dt.strftime('%Y%m')

# 年月とcustomer_idごとに集計した結果をuselog_monthsに代入
uselog_months = uselog.groupby(['年月', 'customer_id'], as_index=False).count()

# log_idをcountにrenameする
uselog_months.rename(columns={'log_id':'count'}, inplace=True)

# usedateを削除
del uselog_months['usedate']
uselog_months.head()
```

```python
# uselog_months['年月']の期間を確認
print(uselog_months['年月'].min())
print(uselog_months['年月'].max())

# uselog_monthsの年月例の固有値のみをリスト化
year_months = list(uselog_months['年月'].unique())

# データフレーム作成
predict_data = pd.DataFrame()

# 2018年10月〜2019年3月を指定（？）
for i in range(6, len(year_months)):
    tmp = uselog_months.loc[uselog_months['年月'] == year_months[i]]
    tmp.rename(columns={'count':'count_pred'}, inplace=True)
    for j in range(1, 7):
        # iで指定した年月から、jを使って6ヶ月前までさかのぼってuselog_monthsの6ヶ月前までの利用データを格納
        tmp_before = uselog_months.loc[uselog_months['年月'] == year_months[i-j]]
        # tmp_beforeの年月カラムを削除
        del tmp_before['年月']
        # format関数で、()で指定した変数（今回はi-j）を文字列{}に埋め込むことができる
        # つまり、tmp_beforeのcount列の名前をcount_0からcount_6まで格納する
        tmp_before.rename(columns={'count':'count_{}'.format(j-1)}, inplace=True)
        # customer_idをもとに、tmpとtmp_beforeを結合
        tmp = pd.merge(tmp, tmp_before, on='customer_id', how='left')
        tmp_before.head()
    # tmpをpredict_dataに結合
    predict_data = pd.concat([predict_data, tmp], ignore_index=True)

predict_data.head()
```

```python
# 欠損値処理
predict_data = predict_data.dropna()

# 欠損値を含んでいた列を削除する
predict_data = predict_data.reset_index(drop=True)
predict_data.head()
```

### ノック37：特徴となる変数を付与しよう

```python
# customer_idをもとに、predict_dataにcustomerテーブルのstart_dateを結合
predict_data = pd.merge(predict_data, customer[['customer_id', 'start_date']], on='customer_id', how='left')
predict_data.head()
```

```python
# 年月をdatetime型に変換し、strftimeで年月形式に変換
predict_data['now_date'] = pd.to_datetime(predict_data['年月'], format='%Y%m')
# start_dateもdatetime型に変換
predict_data['start_date'] = pd.to_datetime(predict_data['start_date'])
# relativedeltaで月の演算をしていく
from dateutil.relativedelta import relativedelta
# 会員期間を算出するため、算出結果を格納するためのperiod列を作って初期化しておく
predict_data['period'] = None
# predict_dataの行数だけ繰り返し処理
for i in range(len(predict_data)):
    # now_dateとstart_dateを減算し、delta変数へ格納
    delta = relativedelta(predict_data['now_date'][i], predict_data['start_date'][i])
    # 月単位に直し、会員期間を算出
    predict_data['period'][i] = delta.years*12+delta.months

predict_data.head()
```

### ノック38：来月の利用回数予測モデルを作成しよう

```python
# 2018年4月以降に入会した人を対象に学習していく
predict_data = predict_data.loc[predict_data['start_date'] >= pd.to_datetime('20180401')]
# ライブラリ
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# インスタンスを立てる
model = LinearRegression()
# 説明変数にpredict_dataのcount_0からperiodを指定
x = predict_data[['count_0', 'count_1', 'count_2', 'count_3', 'count_4', 'count_5', 'period']]
# 目的変数にcount_predを指定
y = predict_data['count_pred']
# 学習データと評価データ分割
X_train, X_test, y_train, y_test = train_test_split(x, y)
# モデル学習
model.fit(X_train, y_train)
```

```python
# 学習データをもとに決定係数を算出
print(model.score(X_train, y_train))
# 評価データをもとに決定係数を算出
print(model.score(X_test, y_test))
```

### ノック39：モデルに寄与している変数を確認しよう

```python
coef = pd.DataFrame({'feature_name':x.columns, 'coefficient':model.coef_})
coef
```
